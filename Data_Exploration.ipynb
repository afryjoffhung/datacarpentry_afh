```python
#Import packages for plotting
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
#Plot life expectancy as a histogram

plt.hist(gapminder_copy['life Exp'])
plt.xlabel('life Exp')
plt.ylabel('count')
```

```python
sns.boxplot(x= "year", y = "life Exp", data = gapminder_copy)
plt.xlabel('Year')
plt.ylabel('Life Expectancy')

#Visualizations are a good way to view trends and outliers
```

```python
plt.scatter(gapminder_copy['gdpPercap'], gapminder_copy ['life Exp'])
plt.xlabel('GDP per cap')
plt.ylabel('Life Expectancy')
```

```python
plt.scatter(gapminder_copy['gdpPercap'], gapminder_copy ['life Exp'])
plt.xscale('log')
plt.xlabel('GDP per cap')
plt.ylabel('Life Expectancy')
plt.savefig('log_plot.png')
```
